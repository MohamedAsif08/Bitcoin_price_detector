```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
```

```python
data = pd.read_csv('bitcoin.csv')
data.head()
```

```python
data.drop('Date', axis =1, inplace = True)
prediction_days = 30
data['Prediction'] = data['Price'].shift(-prediction_days)
data
```

```python
x = np.array(data.drop('Prediction',1))
x = x[:-prediction_days]
y = np.array(data.drop('Price',1))
y = y[:-prediction_days]
print(x ,"\n", y)
```

```python
prediction_day_arrays = np.array(data.drop('Prediction',1)[-prediction_days:])
print(prediction_day_arrays)
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size = 0.2)
svr_rbf = SVR(kernel='rbf', gamma=0.00001, C = 1e3)
svr_rbf.fit(xtrain,ytrain)
```

```python
svr_rbf_accuracy = svr_rbf.score(xtest, ytest)
print("Accuracy :%0.4f"%svr_rbf_accuracy)
```

```python
svr_prediction = svr_rbf.predict(xtest)
print(svr_prediction)
print()
print(ytest)
```

```python
svm_prediction = svr_rbf.predict(prediction_day_arrays)
data['Prediction'][-prediction_days:] = svm_prediction
print(data.tail(prediction_days))
```
